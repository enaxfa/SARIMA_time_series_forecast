```python
import pandas as pd
import numpy as np
from matplotlib import dates
```

```python
df = pd.read_csv('Burger_Sales.csv', index_col = 'Date', parse_dates = True)
```

```python
df.dropna(inplace= True)
```

```python
df.index.max()
```

```python
df.index.min()
```

```python
df['Sumed_Costs'] = df.loc[:, df.columns != 'burger_sales'].sum(axis=1)
```

```python
df
```

```python
ax = df['Sumed_Costs'].resample('MS').mean().plot(title = 'Kretanje troskova oglasavanja');
ax.set(ylabel = 'indijski rupi', xlabel = '');
ax.yaxis.grid(True);
```

```python
ax = df['burger_sales'].resample('MS').mean().plot(title = 'Prodaja burgera');
ax.set(ylabel = 'broj burgera', xlabel = '');
ax.yaxis.grid(True);
```
